```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#only on jupyter
```

```python
import numpy as np
x=np.linspace(0,5,11)    #get total 11 values from 0 to 5 in form of array
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)  #plot x and y values on graph
plt.show()     #show the graph
```

```python
plt.plot(x,y,'#55FF33')     #use color code to change color of data line
plt.show()
```

```python
plt.plot(x,y,'#55FF33')
plt.xlabel('X label')    #label on x axis
plt.ylabel('Y label')    #label on y axis
plt.title('Title')       #title of graph
plt.show()
```

```python
fig=plt.figure()      #plot the graph
axes=fig.add_axes([0.1,0.1,1.3,0.5])    #plot graph on axis(x-axis,y-axis,length,height)  
axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.45,0.15,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')     #give title
axes2.plot(y,x)
axes.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)   #use to make multiple graph
axes[0][0].plot(x,y)
axes[1][1].plot(x,y)
axes[2][2].plot(x,y)
plt.tight_layout()   #make layout more convinie
nt
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title('first')
axes[0][1].plot(x,y)
axes[0][1].set_title('second')
plt.tight_layout()
```
