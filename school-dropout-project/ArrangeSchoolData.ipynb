```python
import pandas as pd
import numpy as np

%matplotlib inline
```

Reading School Data From .xls files

```python
middle_mix=pd.read_excel("data/middle_school_rural_urban.xls")  #get data into new object
middle_mix.isnull().sum()
#middle_mix.describe()
#print(_file_)
```

```python
primary_mix=pd.read_excel("data/primary_rural_urban.xls")
high_school_mix=pd.read_excel("data/9_10_rural_urban.xls")
#df.dropna()
#df.fillna("")
```

merging the data

```python
ndf=primary_mix.merge(middle_mix,on=['STATE/U.T.'])   #merging two files
alldata=ndf.merge(high_school_mix,on=['STATE/U.T.'])

alldata.shape    #shows number of rows and columns

# alldata.to_csv('alldata.csv',index=False)  #used to save the file permanently without common index values  
```

```python
alldata
```

sliceing the data

```python
combined_data=alldata.iloc[-1,:]     #get the last entry with its index values
```

```python
combined_data
```

```python
combined_data.index   #show the index values
```

```python
combined_data.index=[col.strip() for col in combined_data.index]   #removing extra spaces from ends
```

```python
combined_data.index
```

```python
#to remove whitespace in a particular column
combined_data.rename(index={'Enrolment in Class IX  - Girls':'Enrolment in Class IX - Girls'},inplace=True)  
#inplace=True means to change it permanently in memory
```

```python
combined_data.index
```

```python
# initialising an empty dataframe
enrollment=pd.DataFrame(columns=['Boys','Girls','Total'])

#looping over fiffernt grades and gthering data gender wise
classes=['I','II','III','IV','V','VI','VII','VIII','IX','X']
for c in classes:
    boys=combined_data['Enrolment in Class '+c+' - Boys']
    girls=combined_data['Enrolment in Class '+c+' - Girls']
    total=combined_data['Enrolment in Class '+c+' - Total']
    enrollment=enrollment.append({'Boys':boys, 'Girls':girls, 'Total':total}, ignore_index=True)
```

```python
enrollment
```

```python
enrollment.index=np.arange(1,len(enrollment)+1)     #increment index by 1
```

```python
enrollment.plot(title='students enrollend in different classes')
```

```python
enrollment.pct_change()    #percentage of change in data from previous one
```

```python
fraction_dropouts=enrollment.pct_change().dropna() * -1
fraction_dropouts
```

```python
facilities_primary=pd.read_excel('data/facilities_primary.xls')
facilities_secondary=pd.read_excel('data/facilities_secondary.xls')
facilities_u_primary=pd.read_excel('data/facilities_upper_primary.xls')
```

```python
schools_with_toilets=facilities_primary['Availability Within School Premises - Usable Urinal']+facilities_secondary['Availability Within School Premises - Usable Urinal']+facilities_u_primary['Availability Within School Premises - Usable Urinal']

schools_with_toilets.index=facilities_primary['STATE/U.T.']

schools_with_toilets
```

```python
school_category=pd.read_excel('data/school_category.xls')

# subtracting the high school students
#(as we are only analysing trends till class 10)
total_schools=school_category['Total Schools\nof Schools']-school_category['Higher Secondary Schools']

total_schools.index=school_category['STATE/U.T.']
total_schools
```

```python
school_sanitation=pd.DataFrame()
school_sanitation['overall']=total_schools
school_sanitation['with_toilets']=schools_with_toilets
```

```python
school_sanitation
```

```python
school_sanitation['fraction_with_toilets'] =school_sanitation['with_toilets']/school_sanitation['overall']
```

```python
school_sanitation.head()
```

```python
#fraction of girls retaind from class 1st to 10th
girls_retained=alldata['Enrolment in Class X - Girls']/alldata['Enrolment in Class I - Girls']

girls_retained.index=alldata['STATE/U.T.']

school_sanitation['girls_retained']=girls_retained
```
